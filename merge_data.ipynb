```python
import pandas as pd
```

```python
content = pd.read_csv("data/Content (1).csv")
reaction = pd.read_csv("data/Reactions (1).csv")
```

```python
del content["Unnamed: 0"]
del reaction["Unnamed: 0"]
```

```python
reaction = reaction.rename(columns={"User ID": "Reactor ID", "Type": "Reaction Type"})
content = content.rename(columns={"User ID": "Creator ID", "Type": "Content Type"})
```

```python
final_data = reaction.merge(content, on="Content ID").dropna()
```

```python
final_data.to_csv("data/final_data.csv")
```
